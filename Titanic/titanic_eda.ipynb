## Data Structure Observations

- The dataset contains 891 passengers and 12 columns.
- Each row represents one passenger on the Titanic.
- The target variable is `Survived`, where 1 means survived and 0 means did not survive.
- Numerical columns include `Age`, `Fare`, `SibSp`, and `Parch`.
- Categorical columns include `Sex`, `Embarked`, `Cabin`, and `Pclass`.
- The `Age` column has 177 missing values.
- The `Cabin` column has 687 missing values, which is a large portion of the dataset.
- The `Embarked` column has only 2 missing values.

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
(df.isnull().sum()/len(df) *100).sort_values(ascending=False)
```

```python
df["Survived"].value_counts(normalize=True).mul(100).round(2)
```

```python
sns.countplot(data=df, x="Survived")

plt.title("Number of Passengers Who Survived")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")

plt.show()
```

## Survival Distribution

- `0` Represents passengers who did not survive.
- `1` Represents passengers who survived.
- The dataset contains more passengers who did not survive than passengers who survived.
- This is important because `Survived` will be the target variable when building a machine learning model later.
