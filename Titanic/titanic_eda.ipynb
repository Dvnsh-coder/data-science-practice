# Data Cleaning and Machine Learning Preparation

This section creates a cleaned version of the Titanic dataset for machine learning.
The original DataFrame, `df`, is kept unchanged for EDA.

```python
df_ml = df.copy()
```

```python
df_ml.isnull().sum().sort_values(ascending=False   )
```

```python
df_ml["Age"] = df_ml["Age"].fillna(df_ml["Age"].median())
```

```python
df_ml['Age'].isnull().sum()
```

```python
df_ml['Embarked'] = df_ml['Embarked'].fillna(df_ml['Embarked'].mode()[0])
```

```python
df_ml['Embarked'].isnull().sum()
```

```python
df_ml = df_ml.drop(columns = 'Cabin')
```

```python
df_ml.isnull().sum().sort_values(ascending=False   )
```

## Data Cleaning Summary

- Created a separate copy of the original dataset named `df_ml`.
- Filled missing `Age` values using the median age.
- Filled missing `Embarked` values using the most common embarkation port.
- Dropped the `Cabin` column because it contained a large proportion of missing values.
- Kept the original `df` unchanged for exploratory analysis.

# Train-Test Split

The dataset is split into training and testing sets before preprocessing.
This prevents information from the test set from influencing the training process.

```python
model_features = [
    "Pclass",
    "Sex",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "Embarked",
    "FamilySize",
    "IsAlone"
]
```

## Split Summary

- Input features were stored in `X_raw`, and survival outcomes were stored in `y`.
- The dataset was split into 80% training data and 20% testing data.
- `random_state=42` was used to make the split reproducible.
- `stratify=y` was used to preserve a similar survival distribution in the training and testing sets.
- Preprocessing will be fitted on the training data only to prevent data leakage.

# Professional Data Preprocessing Pipeline

This pipeline handles missing values and categorical encoding. It is fitted using training data only to prevent data leakage.


```python
numerical_features = [
    "Pclass",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "FamilySize",
    "IsAlone"
]

categorical_features = [
    "Sex",
    "Embarked"
]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
numerical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median"))
    ]
)
```

```python
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(drop="first", handle_unknown="ignore"))
    ]
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
X_train_prepared = preprocessor.fit_transform(X_train_raw)
```

```python
X_test_prepared = preprocessor.transform(X_test_raw)
```

```python
X_train_prepared.shape
```

```python
X_test_prepared.shape 
```

```python
feature_names = preprocessor.get_feature_names_out()

feature_names
```

## Preprocessing Summary

- Numerical features were processed by filling missing values with the median calculated from the training data.
- Categorical features were processed by filling missing values with the most frequent training value.
- Categorical values were converted into numerical features using one-hot encoding.
- The preprocessing rules were fitted only on training data and then applied to testing data.
- This workflow prevents data leakage.

# Logistic Regression Model

This section builds a machine learning pipeline that preprocesses the data and trains a Logistic Regression classifier to predict Titanic passenger survival.

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", LogisticRegression(max_iter=1000, random_state=42))
    ]
)
```

```python
logistic_model.fit(X_train_raw,y_train)
```

```python
y_pred = logistic_model.predict(X_test_raw)
```
