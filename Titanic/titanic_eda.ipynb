- Numerical columns include `Age`, `Fare`, `SibSp`, and `Parch`.

- The `Embarked` column has only 2 missing values.

- `0` Represents passengers who did not survive.

## Survival Analysis by Gender

This section investigates whether the survival rate differed between male and female passengers.

```python
df["Sex"].value_counts()
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Sex")["Survived"].mean().mul(100).round(2)
```

```python
df.groupby("Sex")["Survived"].agg(['count', 'sum', 'mean'])
```

```python
sns.barplot( data = df, x = 'Sex', y = 'Survived')
plt.title("Survival rate by gender")
plt.xlabel("Gender")
plt.ylabel("Survial Rate")

plt.show()
```

```python
sns.countplot(data=df, x="Sex", hue="Survived")

plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

### Insight

- The survival rate was different for male and female passengers.
- Female passengers had a higher/lower survival rate than male passengers.
- Therefore, `Sex` appears to be an important feature for predicting survival in a future machine learning model.
- This shows an association in this historical dataset; it does not by itself prove causation.

## Survival Analysis by Passenger Class

This section investigates whether passenger class was related to survival.

```python
df['Pclass'].value_counts().sort_index()
```

```python
df.groupby("Pclass")['Survived'].mean().mul(100).round(2)
```

```python
class_summary = df.groupby("Pclass")["Survived"].agg(
    Total_Passengers="count",
    Total_Survived="sum",
    Survival_Rate="mean"
)
class_summary['Survival_Rate'] = class_summary["Survival_Rate"] .mul(100).round(2)
class_summary
```

```python
sns.barplot (data = df , x= 'Pclass', y= 'Survived' )


plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class (1 = First, 2 = Second, 3 = Third)")
plt.ylabel("Average Survival Rate")

plt.show()
```

```python
sns.countplot(data= df, x="Pclass", hue="Survived")


plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

### Insight

- Passenger class appears to have a relationship with survival.
- The survival rates of first-62.96, second-47.28, and third-24.24 class passengers.
- First class has the highest survival rate and Third class has the lowest survival rate.
- `Pclass` is likely to be a useful feature for a future machine learning model.

## Age Analysis and Missing Values

This section explores the age distribution of Titanic passengers, its relationship with survival, and missing values in the `Age` column.

```python
df['Age'].value_counts(normalize='True')*100
```

```python
df['Age'].isnull().sum()
```

```python
(df['Age'].isnull().sum()/len(df)*100).round(2)
```

- The `Age` column has 177 missing values.
- This represents 19.87% of the dataset.
- Missing ages need to be handled before using `Age` in a machine learning mode.

```python
df['Age'].describe()
```

```python
sns.histplot(data=df, x="Age", bins=30, kde=True)

plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.histplot(
    data=df,
    x="Age",
    hue="Survived",
    bins=30,
    kde=True,
    multiple="layer"
)

plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
df.groupby("Survived")["Age"].agg(["count", "mean", "median", "min", "max"])
```

```python
sns.boxplot(data=df, x="Survived", y="Age")

plt.title("Age Distribution by Survival Status")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")

plt.show()
```

```python
df.groupby(["Pclass", "Sex", ])["Age"].median()
```

### Insights

- Most passengers in the dataset were in their 20s and 30s.
- The `Age` column contains missing values and must be handled before machine learning.
- The median age was 28 years for both survivors and non-survivors.
- The age distributions of survivors and non-survivors overlap considerably.
- Age may be useful for predicting survival, but it should be analysed along with other variables such as gender and passenger class.
